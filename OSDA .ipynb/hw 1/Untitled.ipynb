```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

# Датасет 1

```python
df = pd.read_csv("cancer patient data sets.csv", index_col="index")
```

```python
df
```

```python
df.shape
```

```python
df.dtypes
```

```python
for c in df.columns.drop(["Patient Id", "Age"]):
    print(pd.unique(df[c]))
```

```python
raw_df = df[df.columns.drop(["Patient Id", "Age"])]
raw_df
```

```python
ohe = OneHotEncoder(dtype=bool)
```

```python
ohe.fit(raw_df)
```

```python
ohe.categories_
```

```python
res_columns = ohe.get_feature_names_out(raw_df.columns)
res_columns
```

```python
bin_df = ohe.transform(raw_df).toarray()
```

```python
bin_df.dtype
```

```python
bin_df = pd.DataFrame(bin_df, columns=res_columns, index=df.index)
```

```python
bin_df
```

```python
float_df = df[["Patient Id", "Age"]]
float_df
```

```python
res_df = float_df.merge(right=bin_df, on='index')
res_df
```
